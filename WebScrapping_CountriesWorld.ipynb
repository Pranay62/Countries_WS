```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

countries_details = []

url = 'https://www.scrapethissite.com/pages/simple/'
response = requests.get(url)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
# Find all the team elements
countries_elements = soup.find_all(class_='col-md-4 country')
#print(countries_elements)

for element in countries_elements:
    country_name = element.find(class_='country-name').text.strip()
    country_capital = element.find(class_='country-capital').text.strip()
    country_population = element.find(class_='country-population').text.strip()
    country_area = element.find(class_='country-area').text.strip()    
    countries_details.append({'Country Name':country_name,'Capital':country_capital,'Population':country_population,'Area (km2)':country_area})
    
# Create a DataFrame from the team details
df = pd.DataFrame(countries_details)

# Print the DataFrame
df
```
